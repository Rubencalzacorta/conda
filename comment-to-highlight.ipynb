```python
#====PASTE THE "meeting_saved_chat.txt" in the same folder====

f = open("meeting_saved_chat.txt", "r")
allText = []
for line in f:
    allText.append(line)
    
```

```python
# ==============Supporting Function - NO INPUT=================
import re 

def get_sec(time_str):
    """Get Seconds from time."""
    if len(time_str.split(':')) > 2 :
        h, m, s = time_str.split(':')
        return int(h) * 3600 + int(m) * 60 + int(s)
    elif len(time_str.split(':')) == 2:
        m, s = time_str.split(':')
        return int(m) * 60 + int(s)
    else : 
        return int(time_str)
    
```

```python
# ====================== Input meetingStartTime =================

# meeting start time input manually or with -start command   , input manually timeMark

meetingStartTimeInputHHMMSS = "00:00:00"
keywordToStart = "-start" 
timeMark = 1586416622571
# all commands that will generate a highlight ##notify ruben to inclue or exclude any
currentCommands = ("-start", "-", "--", "—", "++", "+")

# ====================== Input meetingStartTime =================
```

```python
#======================= NO INPUT NEEDED ========================

strapiObject = []
meetingStartTime = get_sec(meetingStartTimeInputHHMMSS)
# preworking the text lines so every line is a comment and not an individual line. 
lastComment = 0
lineToDelete = []

for idx, line in enumerate(allText):
    
    if re.match(r'^(?:[01]\d|2[0123]):(?:[012345]\d):(?:[012345]\d)', line):
        lastComment = idx
    else : 
        allText[lastComment] = allText[lastComment] + " " + line
        lineToDelete.append(idx)

# now deleting all extra lines
for index in sorted(lineToDelete, reverse=True):
    del allText[index]


# checking if lines are highlights or just comments. current commands will show if it is or is now a highlight
for line in allText:

    commentText = line.split(" : ", 1)[1].strip()
    
    if commentText.startswith(currentCommands):
        userName = re.split('From | : ',line)[1]
        timeHHMMSS = line.split(" ")[0] 
        startTime = 0

        if commentText.startswith(keywordToStart):
            meetingStartTime = get_sec(timeHHMMSS)

        elif commentText.startswith("—"):
            startTime = get_sec(timeHHMMSS)-meetingStartTime - 60
            commentText = commentText[2:].strip().capitalize()   

        elif commentText.startswith("-"):
            startTime = get_sec(timeHHMMSS)-meetingStartTime - 30
            commentText = commentText[1:].strip().capitalize()   

        elif commentText.startswith("++"):
            startTime = get_sec(timeHHMMSS)-meetingStartTime + 60
            commentText = commentText[2:].strip().capitalize()  

        elif commentText.startswith("+"):
            startTime = get_sec(timeHHMMSS)-meetingStartTime + 30
            commentText = commentText[1:].strip().capitalize()  
        
#       starttime different to in all cases except fot he -start command
        if startTime != 0:
            strapiObject.append({
                      "userName": userName,
                      "word": commentText,
                      "time": timeMark,
                      "startTime": startTime,
                      "duration": 1
                       })

```

```python
#==============No Input Needed ========================

# checks and results
import time

# create a file with writting permits
newFile = open("strapiFile.txt","w+")
newFile.write(f'there where {len(strapiObject)} entries during the meeting \r\n\n\n')
print(f'there where {len(strapiObject)} entries during the meeting \r\n\n\n')


entryIndex = 1
# extract words from entries
segmentsCheck = []
for entry in strapiObject:
             segmentsCheck.append(entry['word'])
# extract time in hhmmss format from entries        
timeCheck = []
for entry in strapiObject:
             timeCheck.append(time.strftime('%H:%M:%S', time.gmtime(entry['startTime'])))

# write in document readble forms of the entry for a manual check
for entry in segmentsCheck:
    newFile.write(f'{entryIndex}. -{timeCheck[entryIndex-1]}- {entry} \n')
    print(f'{entryIndex}. -{timeCheck[entryIndex-1]}- {entry} \n')
    entryIndex += 1

# write in doc the Json for strapi
newFile.write(f'\n\n {strapiObject}')
print(f'\n\n {strapiObject}')

# close file
newFile.close() 



```
